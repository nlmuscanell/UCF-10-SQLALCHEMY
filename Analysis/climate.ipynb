## Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
# Reflect an existing database into a new model
engine = create_engine("sqlite:///../Data/hawaii.sqlite")
Base = automap_base()

# Reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()

# Could also use inspector.get_tablenames
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(bind=engine)
```

```python
### Temperature Analysis I

# Hawaii is reputed to enjoy mild weather all year. 

# Is there a meaningful difference between the temperature in, for example, June and December?

# You may either use SQLAlchemy or pandas's `read_csv()` to perform this portion.

# Identify the average temperature in June at all stations across all available years in the dataset. 

# Do the same for December temperature.

# Use the t-test to determine whether the difference in the means, if any, is statistically significant. 
# Will you use a paired t-test, or an unpaired t-test? Why?
```

```python
# Reading CSV File
df = pd.read_csv("/Users/nicolemuscanell/UCF-10-SQLAlchemy/Data/hawaii_measurements.csv")
df.head()
```

```python
# Creating DF
measurement_df = pd.DataFrame(df, columns=['station', 'date', 'prcp', 'tobs'])
measurement_df.head()
```

```python
print(measurement_df.dtypes)
```

```python
# Converting dates
measurement_df['date'] = pd.to_datetime(measurement_df['date'], format='%Y-%m-%d')
measurement_df.head()
```

```python
print(measurement_df.dtypes)
```

```python
# Creating DF for june temperatures
june_data = measurement_df.loc[(measurement_df['date'].dt.month==6)]
june_data.head()
```

```python
# Creating DF for december temperatures
december_data = measurement_df.loc[(measurement_df['date'].dt.month==12)]
december_data.head()
```

```python
# Calculate the average temp for june
june_avg = june_data['tobs'].mean()
print(f"The average temperature during June was {round(june_avg,1)}.")
```

```python
# Calculate the average temp for december
dec_avg = december_data['tobs'].mean()
print(f"The average temperature during December was {round(dec_avg,1)}.")
```

```python
# Conduct independent samples t-test
stats.ttest_ind(june_data.tobs, december_data.tobs, equal_var=False)
```

**Observation**: The average temperature across all stations during June was 74.9 degrees, whereas the average temperarure during December was 71 degrees. This difference is statistically significant, p < .001.

### For the grader:

It doesn't really make sense to plan a trip that was in the past and for which there is already existing data, so I am simply using the latest data (2017) to plan my hypothetical trip that would be in 2021.

```python
# Calculate the min, avg, and max temperatures for your trip
# My trip dates: start (2021-07-20), end (2021-07-26)
# Final dates (matching dates from last available year): start (2017-07-20), end (2017-07-26)

def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

print(calc_temps('2017-07-20', '2017-07-26'))
```

```python
# Storing the min, avg, and max
trip = calc_temps('2017-07-20', '2017-07-26')
trip
```

```python
# Storing into a DF
trip_data = pd.DataFrame(trip, columns=["TMIN", "TAVG", "TMAX"])
trip_data
```

```python
# Bar plot for trip average temp

# Define error
error = trip_data['TMAX']-trip_data['TMIN']

# Plot
trip_data.plot.bar(y="TAVG", yerr=error, color="mediumslateblue", edgecolor='k', 
                   alpha=0.6, legend=None, figsize=(5,7))
plt.title('Trip Average Temperature', fontsize=14, fontweight='bold')
plt.xticks([])
plt.ylabel("Temperature (Fahrenheit)", fontsize=14, fontweight='bold')
plt.ylim(0, 100, 20)
plt.tight_layout()
```

### Bonus - Daily Normals

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip

# Use the start and end date to create a range of dates

# Stip off the year and save a list of %m-%d strings

# Loop through the list of %m-%d strings and calculate the normals for each date
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# Set the start and end date of the trip and strip off the year
start_date = date(2017, 7, 20)
end_date = date(2015, 7, 26)
number_of_days=7

# Use the start and end date to create a range of dates and save to a list
date_list = []

for day in range(number_of_days):
  a_date = (start_date + dt.timedelta(days = day)).isoformat()
  date_list.append(a_date)
    
print(date_list)
```

```python
# Turn dates from string to datetime
date_list =[dt.datetime.strptime(date, "%Y-%m-%d") for date in date_list]
date_list
```

```python
# Save a list of %m-%d strings
trip_dates =[dt.date.strftime(date, "%m-%d") for date in date_list]
print(trip_dates)
```

```python
normals_list = []

def daily_normals(date):
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
for date in trip_dates:
    #daily_normals(date).append(normals_list)
    normals_list.append(daily_normals(date)[0])
    
normals_list
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
daily_normals = pd.DataFrame(normals_list, columns=["tmin","tavg", "tmax"], index=trip_dates)
daily_normals
```

```python
# Plot the daily normals as an area plot with `stacked=False`
daily_normals.plot.area(stacked=False)
plt.title("Daily Normals for July 20 to July 26th", fontsize=14, fontweight='bold')
plt.xlabel("Date", fontsize=14, fontweight='bold')
plt.ylabel("Temperature (Fahrenheit)", fontsize=14, fontweight='bold')
plt.xticks([])
plt.show()
```
